```python
import matplotlib.pyplot as plt
```

```python
# Ler o conteúdo dos arquivos
with open('Gloss1018.json', 'r') as f1, open('Dloss1018.json', 'r') as f2:
    dados1 = eval(f1.read())
    dados2 = eval(f2.read())

# Ler o conteúdo dos novos arquivos a serem adicionados
with open('Gloss1517.json', 'r') as nf1, open('Dloss1517.json', 'r') as nf2:
    novos_dados1 = eval(nf1.read())
    novos_dados2 = eval(nf2.read())

# Ler o conteúdo dos novos arquivos a serem adicionados
with open('Gloss2016.json', 'r') as rf1, open('Dloss2016.json', 'r') as rf2:
    novos_dados3 = eval(rf1.read())
    novos_dados4 = eval(rf2.read())

# Concatenar as listas de dados
dados1 += novos_dados1
dados2 += novos_dados2

dados1 += novos_dados3
dados2 += novos_dados4
```

```python
# Criar uma figura com tamanho personalizado
plt.figure(figsize=(15, 6))  # Largura: 10 polegadas, Altura: 6 polegada

# Criar um gráfico de linha
indice_inicial = 500
indices = range(indice_inicial, indice_inicial + len(dados1))
indices = range(indice_inicial, indice_inicial + len(dados2))
plt.plot(indices, dados2, label='Discriminador')
plt.plot(indices, dados1, label='Gerador')


# Adicionar rótulos e título
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Gráfico da loss')

# Adicionar legenda
plt.legend()

# Exibir o gráfico
plt.show()
```
